```python
# Let's make an empty array with the same size as the image array
responsivity_PI640_BLUE_array_fit_substracted = np.zeros_like(responsivity_PI640_BLUE_array_fit)
# Let's also get the height and the width of the image array
Height = np.shape(responsivity_PI640_BLUE_array_fit)[0]
Width = np.shape(responsivity_PI640_BLUE_array_fit)[1]
# This is where the horizontal bars are located, which we find out in the cells below 
Horizontal_bar_heights = [14,50,88, 125, 163, 200, 237, 275]
# Go through the height and the width of the image, i.e. all the pixels
for height in range(Height):
    for width in range(Width):
        # Get the value of the current pixel
        current_pixel_value = responsivity_PI640_BLUE_array_fit[height][width]
        # If the value is less than 190
        if current_pixel_value < 190:
            # If we are at the heights of the horizontal bars
            if height in Horizontal_bar_heights:
                # Then we check whether the current pixel is on the curve or on the horizontal lines,
                # If it is on the horizontal bars, then either the pixel above or below it will be a white pixel (white value),...
                # Because we noticed above that the horizontal bars are at most 2 pixels thick in height
                # If both pixels are less than 200 (not white), the current pixel is on the curve (where they also meet the horizontal bars)
                if ((responsivity_PI640_BLUE_array_fit[height-1][width] < 200) and (responsivity_PI640_BLUE_array_fit[height+1][width] < 200)):
                    # Then we record the pixel value
                    responsivity_PI640_BLUE_array_fit_substracted[height][width] = current_pixel_value

                # Otherwise don't record it because it is on the horizontal bar, already done by the if condition, Don't need else (on this 2nd inner if)

            else:
                # Of the outer if condition, then we are not at the height where the horizontal bars are, so the pixel is on the curve
                # So, we record it
                responsivity_PI640_BLUE_array_fit_substracted[height][width] = current_pixel_value
```

```python
# Let's plot the image, which now only has the curve in it (now appearing bright because is a non zero value)
# while the rest is recorded as a zero value
fig, ax = plt.subplots(figsize=(20, 20))

# Display the image
im = ax.imshow(responsivity_PI640_BLUE_array_fit_substracted, cmap='viridis')

# Add colorbar
cbar = plt.colorbar(im)

# Add labels and title
ax.set_xlabel('Width')
ax.set_ylabel('Height')
ax.set_title('Blue channel')

# Show the plot
plt.show()
```

```python
# First let's make them into tuples so that we don't lose the pixel location
responsivity_PI640_BLUE_array_fit_substracted_curve_tuple_s = []
# The second of the shape gives the length of arrays
for i in range(np.shape(responsivity_PI640_BLUE_array_fit_substracted_curve)[1]):
    responsivity_PI640_BLUE_array_fit_substracted_curve_tuple_s.append((responsivity_PI640_BLUE_array_fit_substracted_curve[0][i], responsivity_PI640_BLUE_array_fit_substracted_curve[1][i]))
```

```python
# Now the location of a pixel on the curve is given by (height, width) pairs
responsivity_PI640_BLUE_array_fit_substracted_curve_tuple_s
```

```python
# Let's use a copy of the above list of locations of pixels, because we will set some data to zeros
responsivity_PI640_BLUE_array_fit_substracted_curve_tuple_s_copy = np.copy(responsivity_PI640_BLUE_array_fit_substracted_curve_tuple_s)
# nodup means no duplicates of the wavelength which is the width pixels axis, which is the second of the tuples.
responsivity_PI640_BLUE_array_fit_substracted_curve_tuple_s_nodup =  []
# For the size of the list
for i in range(np.shape(responsivity_PI640_BLUE_array_fit_substracted_curve_tuple_s_copy)[0]):
    # Get a value on one of the pixels
    current_total = responsivity_PI640_BLUE_array_fit_substracted_curve_tuple_s_copy[i][0]
    # current width that represents the wavelength
    current_width_wavelength  = responsivity_PI640_BLUE_array_fit_substracted_curve_tuple_s_copy[i][1]
    # Number of duplicates at this width
    current_number_dups = 1
    # Check only after the current one in the list for duplication
    for j in range(i+1, np.shape(responsivity_PI640_BLUE_array_fit_substracted_curve_tuple_s_copy)[0], 1):
        # If the width are equal, meaning if there exist multiple values at the same wavelength (width)
        if responsivity_PI640_BLUE_array_fit_substracted_curve_tuple_s_copy[i][1] == responsivity_PI640_BLUE_array_fit_substracted_curve_tuple_s_copy[j][1]:
            # Add the value of the heights
            current_total += responsivity_PI640_BLUE_array_fit_substracted_curve_tuple_s_copy[j][0]
            # 1 more duplicate (same value, the first was counted as 1) found
            current_number_dups += 1
            
            # Now set the ones I already counted to 0 (could detele but the loop index would go out of range)
            # so that I don't run into them again when going down
            responsivity_PI640_BLUE_array_fit_substracted_curve_tuple_s_copy[j][0] = 0
            # Setting the height to zero isn't really necessary but let's do it anyway
            responsivity_PI640_BLUE_array_fit_substracted_curve_tuple_s_copy[j][1] = 0
    # Take the average
    current_average = current_total / current_number_dups
    ####################################
    #### WARNING #######################
    ####################################
    # Now width is first (wavelength)
    # Height is second (responsivity)
    
    responsivity_PI640_BLUE_array_fit_substracted_curve_tuple_s_nodup.append((current_width_wavelength, current_average))
```

```python
# Let's get rid of the zeros
responsivity_PI640_BLUE_array_fit_substracted_curve_pixels = []
for i in range(np.shape(responsivity_PI640_BLUE_array_fit_substracted_curve_tuple_s_nodup)[0]):
    if responsivity_PI640_BLUE_array_fit_substracted_curve_tuple_s_nodup[i][0] != 0:
        # Append the tuples whose first entry (wavelength) isn't zero (set for duplicates earlier)
        responsivity_PI640_BLUE_array_fit_substracted_curve_pixels.append((responsivity_PI640_BLUE_array_fit_substracted_curve_tuple_s_nodup[i][0], responsivity_PI640_BLUE_array_fit_substracted_curve_tuple_s_nodup[i][1]))
```

```python
# Let's sort the list by wavelength, which now is the first element of the tuple
responsivity_PI640_BLUE_array_fit_substracted_curve_pixels = sorted(responsivity_PI640_BLUE_array_fit_substracted_curve_pixels, key=lambda t: t[0])
```

```python
# Let's check if any pixel is missing on the wavelength
for i in range(4, 388):
    if responsivity_PI640_BLUE_array_fit_substracted_curve_pixels[i-4][0] != i:
        print(f'found once, {i}')
# Nothing was printed, this means the wavelengths pixels are continuous,
# and there are no gaps in the middle, hence they are also uniformly distributed
```

```python
# Now based on the location of the markings on the scale and the value of the first pixel,
# Let's convert the pixel data into (wavelength, responsivity) data
responsivity_PI640_values =  []
for i in range(np.shape(responsivity_PI640_BLUE_array_fit_substracted_curve_pixels)[0]):
    # Also multiply by 10e-6 to account for the fact this is in micrometers
    current_wavelength = ((0.0267 * responsivity_PI640_BLUE_array_fit_substracted_curve_pixels[i][0]) + 6.5849) * 10**(-6)
    # Because we are starting from the top, we substract
    current_responsivity = (0.8378 - (0.00270 * responsivity_PI640_BLUE_array_fit_substracted_curve_pixels[i][1]))
    responsivity_PI640_values.append((current_wavelength, current_responsivity))
```

#### b. Redefine integrator_with_emissivity_term_s_Matrix_w_responsivity

```python
def integrator_with_emissivity_term_s_Matrix_w_responsivity(T_couple_s, responsivity_tuples, Emissivity_terms):
    '''
    Arguments:
    1. T_couple_s : This is the list of thermocouple temperature data at each frame, and is close to the directly read data  from a thermocouple
    
    2. responsivity_tuples: a list of tuples giving the responsivity
        where the first element in the tuple is a wavelength and the second element is the responsivity at that wavelength
        For example: [(4.00 x10e-6 meters, 0.1), (4.01 x10e-6 meters, 0.11), (4.02 x 10e-6 meters, 0.22), ...]

    In this case the information about  the first and the last wavelength for the range of integration, 
    and the number of integration points will be obtained from the wavelengths available in the argument responsivity_tuples

    ---- So, that we can use the integrated_B_non_Blackbody_w_responsivity function above.
    
    3.
    Emissivity_terms: For example, [lambda, T, lambda * T]
    It is a list whose each entry...
    ---- is an anonymous function that represents a term of the emissivity function. 

    Returns: 
    A matrix
    whose each row represents values at a Temperature, corresponding to the temperature in T_couple_s at that position in the list
    whose each column represents values for a term of the emissivity function
    and the entries at each cell (row i, column j) are the integrated intensities...
    --- at temperature T_i (ith temperature)
    --- with the jth emissivity term of the emissivity function
    
    '''
    # Get the number of terms of the emissivity function
    number_of_emissivity_terms = np.size(Emissivity_terms)
    
    # Define a matrix of zeros to hold the total integrated intensity at each temperature, ...
    # Here each column is for one temperature
    # In each column is the contribution to intensity weighed by one term of the emissivity function 
    Matrix_integrated_non_blackbody_intensity = np.zeros((np.size(T_couple_s), number_of_emissivity_terms))

    
    # For each term of the emissivity model
    for i in range(number_of_emissivity_terms):
        # Get the integrated B with emissvity using the integrated_B_non_Blackbody function in section 4.1 ...
        # put it into the ith column of the Matrix
        Matrix_integrated_non_blackbody_intensity[:,i] = np.array(integrated_B_non_Blackbody_w_responsivity(T_couple_s, responsivity_tuples, Emissivity_terms[i]))
        
        print(f'{i+1} of {number_of_emissivity_terms} done.')
    return Matrix_integrated_non_blackbody_intensity
```

```python
# Let's make this into a function
def integrated_B_non_Blackbody_w_responsivity(T_couple_s, responsivity_tuples, emissivity_of_T_couple_then_lambda):
    '''
    Arguments:
    1. T_couple_s : This is the list of thermocouple temperature data at each frame, and is close to the directly read data  from a thermocouple

    2. responsivity_tuples: a list of tuples giving the responsivity
        where the first element in the tuple is a wavelength and the second element is the responsivity at that wavelength
        For example: [(4.00 x10e-6 meters, 0.1), (4.01 x10e-6 meters, 0.11), (4.02 x 10e-6 meters, 0.22), ...]

    In this case the information about  the first and the last wavelength for the range of integration, 
    and the number of integration points will be obtained from the wavelengths available in the argument responsivity_tuples

    ---- So, that we can use the trapezoidal_multiplier_w_responsivity function above.
    
    3. emissivity_of_T_couple_then_lambda: is an emissivity function of Temperature that returns a function of the wavelength at that temperature
    This is theoretically an emissivity of a temperature and wavelength, where the temperature is evaluated first and then wavelength 

    Returns: 
    B_lambda_T_s_integrated: A list of The intensity at each temperature, for a non black body; since here we multiply by emissivity and responsivity
    
    Description:
    This function passes in intensity at a certain temperature and the emissivity at the same temperature to the trapezoidal_multiplier_w_emissivity ...
    function where (intensity_at_T_at_lambda * emissivity_at_T_at_lambda * responsivity_at_lambda) d lambda, will be integrated ...
    for each temperature in T_couple_s
    
    Just like integrated_B_blackbody function integrates the intensity using trapezoidal function
    this function integrates the intensity for a non black body which is a product of the intensity for blackbody and emissivity, using ...
    trapezoidal_multiplier function
    '''
    
    # Get the intensity as functions of wavelengths at various temperatures
    T_couple_s_at_T = B_lambda_T_s_at_T(T_couple_s)
    # Define an empty list to hold the integrated intensity values
    B_lambda_T_s_integrated = []
    
    # For each of the functions of wavelengths at various temperatures
    for i in range(np.size(T_couple_s_at_T)):
        # Get the emissivity at this temperature, as a function of the wavelength
        # This means evaluate the emissivity at this temperature
        emissivity_at_T = emissivity_of_T_couple_then_lambda(T_couple_s[i])
        # Append the expression for blackbody intensity and the emissivity to the list of functions to be multiplied
        # Both are functions of wavelength at this temperature
        functions_to_be_multiplied = [T_couple_s_at_T[i], emissivity_at_T]
        # Call the trapezoidal function and append the result of integration at this temperature to the B_lambda_T_s_integrated array
        B_lambda_T_s_integrated.append(trapezoidal_multiplier_w_responsivity(functions_to_be_multiplied, responsivity_tuples))
        #### Just need to pass the responsivity tuples to trapezoidal_multiplier_w_responsivity, where the multiplication will be done

    return B_lambda_T_s_integrated
```

```python
# We will first define a slight variation of the trapezoidal_multiplier function that multiplies the values of ...
# the intensity of a blackbody, the emissivity and the responsivity, at wavelengths of interpolation x_k ...
# So that it does integration of: Blackbody intensity * emissivity * responsivity
def trapezoidal_multiplier_w_responsivity(f_s, responsivity_tuples):
    '''
    Arguments:
    f_s : a list of functions of wavelength. The model to have in mind for now is [B_of_lambda, epsilon_of_lambda]
    responsivity_tuples: a list of tuples giving the responsivity
        where the first element in the tuple is a wavelength and the second element is the responsivity at that wavelength
        For example: [(4.00 x10e-6 meters, 0.1), (4.01 x10e-6 meters, 0.11), (4.02 x 10e-6 meters, 0.22), ...]

    In this case the information about  the first and the last wavelength for the range of integration, 
    and the number of integration points will be obtained from the wavelengths available in the argument responsivity_tuples

    Retuns:
    The integration of :B_of_lambda * epsilon_of_lambda * responsivity, d_wavelength 
    by using interpolated values of B_of_lambda(x_k) * epsilon_of_lambda(x_k) * responsivity (x_k) for x_k being the wavelengths used for integration
    which are the linspace approximations of the ones obtained from the argument responsivity_tuples
    
    Description:
    This function is a slight adjustment of the trapezoidal_multipler function from Section 4.1
    '''
    # The wavelength of the first tuple is the lower integration limit wavelength
    x_0 = responsivity_tuples[0][0]
    # The wavelength of the last tuple is the uppper integration limit for the wavelength
    x_N = responsivity_tuples[-1][0]
    # Remember we need {x_0, ..., x_n}, so N+1 many values for the integration
    # So, we say we have N+1 values from responsivity_tuples, so N = the list length - 1

    # We are not creating a list of wavelengths, but just using the responsivity values directly
    # How the value of N is, is supposed to uniformly distribute the wavelengths,
    # which they are more or less so because they were read from 
    N = np.shape(responsivity_tuples)[0] - 1
    
    
    delta_x = float(x_N - x_0) / N
    sum = 0.0

    # For x_0
    prod = 1.0
    # For each of the functions, multiply the value of the functions at x_0 together,
    # Then add that value to the sum as the trapezoidal function would do.
    for one_such_f in f_s:
        prod = prod * (one_such_f(x_0))
    # Multiply the responsivity outside the above loop, so that the mulitplication by responsivity is only done once
    prod = prod  * responsivity_tuples[0][1]
    sum += prod/2.0

    # For x_1 to x_{N-1}
    for i in range(1, N):
        x_k = x_0 + i * delta_x
        
        prod = 1.0
        for one_such_f in f_s:
            prod = prod * (one_such_f(x_k))
        # Multiply the responsivity outside the above loop, so that the mulitplication by responsivity is only done once
        prod = prod * responsivity_tuples[i][1]
        sum += prod

    # For x_N
    prod = 1.0
    for one_such_f in f_s:
        prod = prod * (one_such_f(x_N))
    # Multiply the responsivity outside the above loop, so that the mulitplication by responsivity is only done once
    prod = prod * responsivity_tuples[-1][1]
    sum += prod/2.0
    
    sum = sum * delta_x
    
    return sum
```

```python
# Let's define a function that takes in a list of temperatures and returns a list of functions of wavelength ... 
# which are expressions for the Intensity: B,  at that temperature, which are ready to be plugged into during the integration

def B_lambda_T_s_at_T(some_T_couple_s):
    '''
    Arguments:
    some_T_couple_s : List of temperatures, measured by a thermocouple

    Returns: 
    B_lambda_T_s : a list of anonymous functions, which applied to value of a wavelength, return the intensity B at that wavelength,
                   the list is indexed for each of the temperature passed in, in the list some_T_couple_s

    Description:
    This function takes in a list/numpy array of thermocouple temperatures at various frames: some_T_couple_s
    and returns a list (at/for each temperature in the list some_T_couple_s) of 
    lambda/anonymous functions that take in a wavelength and give you the value of B_lambda_T at that wavelength

    i.e. They plug in the temperature and give a function that has the wavelength yet to be plugged in.

    When this list of lambda/anonymous functions are later passed to the integrator functions like trapezoidal,
    they will use these functions to get the value of B_lambda_T at each wavelength they need.
    '''

    # Create an empty list to hold the anonymous functions/ expressions for the Intensity, B in terms of the wavelength
    B_lambda_T_s = []
    # For each temperature 
    for i in range(np.size(some_T_couple_s)):

        ########################################################################
        ####### WARNING! #######################################################
        ####### Be careful with the lambda/anonymous functions when taking variable in the loop #############
        ####### pass it as a default argument ##################################
        ########################################################################

        # At the given temperature, build the expression for the intensity and append it to the list B_lambda_T_s
        B_lambda_T_s.append(lambda a_lambda, i=i: ( (c_1_lambda / (a_lambda **5)) * (1 / (np.exp(c_2_lambda / (a_lambda * some_T_couple_s[i])) - 1) ) ) )
                            
    return B_lambda_T_s
```

```python
def gradient_descent_without_Constant_MSE_only(N_epochs, T_step_size_MSE, scale_factor_initial_a_J, Integrated_Intensities_Solve, Thermal_Camera_Solve):
    '''
    WARNING:
            Thermal_Camera_Solve needs to be 1 dimensional vector, not a matrix.
            Using emissivity function with 1 term gives a matrix that needs to be flattened earlier
            or by modifying a line in the function here
    
    '''
    
    
    # Get the number of coefficients for the emissivity model
    N_coeffs = np.shape(Integrated_Intensities_Solve)[1]
    N_frames = np.shape(Integrated_Intensities_Solve)[0]
    
    # Because the thermal camera data is (N_frames, 1) matrix (because it was integrated for a list of 1 emissivity function)
    # Let's convert it into a vector and use it like so
    
    # Don't need this because we have flatten it already
    #Thermal_Camera_Solve = Thermal_Camera_Solve.flatten()
    
    # Initialize the coefficients according to the length
    a_J_MSE = scale_factor_initial_a_J * np.ones((N_coeffs,))
    
    # Random initialization did not work very well
    # Let's instead use random initialization
    # a_J_MSE_initial = scale_factor_initial_a_J * np.random.rand(N_coeffs)
    # a_J_MSE = a_J_MSE_initial
    
    # Initialize the error values for all epochs
    MSE_s = np.zeros((N_epochs,))

    # Run the epochs
    for epoch in range(N_epochs):
        # Initialize the errors for this epoch
        MSE_epoch = 0.0

        # Initialize the gradients for this epoch
        MSE_gradient_for_all_coeffs_epoch = np.zeros((N_coeffs,))

        for i in range(N_frames):
            # 1. Get the predicted data- Forward Propagation
            yhat_i_MSE = np.dot(a_J_MSE, Integrated_Intensities_Solve[i])

            # 2. Get the errors
            # Get the contribution from the frame i         
            MSE_i = (1/N_coeffs) * ( (Thermal_Camera_Solve[i] - yhat_i_MSE)**2)
            # Add the mean squared error contribution from the ith frame to the total mean squared error of this epoch
 
            MSE_epoch = MSE_epoch + MSE_i

            # 3. Get the gradients
            # Get the contribution to the gradient from the current frame i
            MSE_gradient_i_for_all_coeffs = (-2/N_coeffs) * (Thermal_Camera_Solve[i] - yhat_i_MSE) * Integrated_Intensities_Solve[i]
            
            # Add the contribution to the gradient from the current frame i to the gradient of this epoch
            MSE_gradient_for_all_coeffs_epoch = MSE_gradient_for_all_coeffs_epoch + MSE_gradient_i_for_all_coeffs

        # 4. Do gradient descent- Backward propagation
        # Calculate the updated coefficients
        if epoch < 10 or epoch > N_epochs-10:
            print(f'a_J_MSE before: {a_J_MSE}')
        
        a_J_MSE = a_J_MSE - (T_step_size_MSE * MSE_gradient_for_all_coeffs_epoch)
        
        #if epoch%1000 ==0 and epoch<N_epochs-3001:
        if epoch%1000 ==0 and epoch<N_epochs-1:   
            current_average = np.mean(a_J_MSE)
            random_noise = np.random.rand(N_coeffs)
            a_J_MSE = a_J_MSE + (0.1 * current_average * random_noise) 
            
        if epoch < 10 or epoch > N_epochs-10:
            print(f'MSE_gradient_for_all_coeffs_epoch: {MSE_gradient_for_all_coeffs_epoch}')
            print(f'T_step_size_MSE * MSE_gradient_for_all_coeffs_epoch: {T_step_size_MSE * MSE_gradient_for_all_coeffs_epoch}')

        # 5. Record the losses of the current epoch            
        MSE_s[epoch] =  MSE_epoch

        if epoch%100 == 0:
            print(f'Epoch {epoch} finished')

    return MSE_s, a_J_MSE
```

```python
def Solve_Test_divider_gradient_descent_with_indices(Thermal_camera_data, Normalized_Intensities, p):
    # p is between 0 and 1
    # p = 0.6 means that for any given frame it has 60 percent probability that it will be sampled for Solving
    # So on average about 60% of the frames should be put in the Solve Set
    N_frames = np.shape(Thermal_camera_data)[0]

    Thermal_camera_Solve = []
    Thermal_camera_Test = []

    Normalized_Intensities_Solve = []
    Normalized_Intensities_Test = []

    Solve_Frames = []
    Test_Frames = []
    
    # Both should have the same number of frames.
    assert(np.shape(Thermal_camera_data)[0] == np.shape(Normalized_Intensities)[0])

    for i in range(N_frames):
        random_value = np.random.random()
        if random_value < p:
            Solve_Frames.append(i)
            Thermal_camera_Solve.append(Thermal_camera_data[i])
            Normalized_Intensities_Solve.append(Normalized_Intensities[i,:])
        else:
            Test_Frames.append(i)
            Thermal_camera_Test.append(Thermal_camera_data[i])
            Normalized_Intensities_Test.append(Normalized_Intensities[i,:])

    Thermal_camera_Solve = np.array(Thermal_camera_Solve)
    Normalized_Intensities_Solve = np.array(Normalized_Intensities_Solve)
    Thermal_camera_Test = np.array(Thermal_camera_Test)
    Normalized_Intensities_Test = np.array(Normalized_Intensities_Test)

    return Thermal_camera_Solve, Normalized_Intensities_Solve, Thermal_camera_Test, Normalized_Intensities_Test, Solve_Frames, Test_Frames
```

```python
def Emissivity_terms2_symbolic(T_couple):
    
    '''
    
    Returns:
    Emissivity_terms2_symbolic: a list of lambda functions that take in the wavelength and return an expression in T_couple
    '''
    # Define some emissivity terms
    ###############################
    #### NOTE:
    # We divide the temperature by 1000, they are in the range 275 Kelvin to around 1500 and 
    # multiply the wavelengths by 10**(6), they are in microwave range which is 10**(-6) meters
    # so that the values don't get too large when we take exponentials.

    # The emissivity function 2
    Emissivity_terms2_symbolic = []

    # 0. Constant term #### 1 term until now ####
    emissivity_constant_term = lambda lambda_: 1.0 * (T_couple/T_couple)
    Emissivity_terms2_symbolic.append(emissivity_constant_term)

    # 1. lambda polynomial terms #### 5 + 1 = 6 terms until now ####
    N_lambda_polynomial = 5
    for i in range(1,N_lambda_polynomial+1):
        current_emissivity_term = lambda lambda_, i=i: ((lambda_*(10**5))**i ) * (T_couple/T_couple)
        Emissivity_terms2_symbolic.append(current_emissivity_term)

    # 2. T polynomial terms #### 2 + 6 = 8 terms until now ####
    N_T_couple_polynomial = 2
    for i in range(1,N_T_couple_polynomial+1):
        current_emissivity_term = lambda lambda_, i=i: (T_couple/1000)**i
        Emissivity_terms2_symbolic.append(current_emissivity_term)
          
    # 3. lambda * T polynomial terms #### 2 + 8 = 10 terms until now ####
    N_lambda_T_couple_polynomial = 2
    for i in range(1,N_lambda_T_couple_polynomial+1):
        current_emissivity_term = lambda lambda_, i=i: ((lambda_*(10**5))*(T_couple/1000))**i
        Emissivity_terms2_symbolic.append(current_emissivity_term)

    # 4. lambda / T polynomial terms #### 2 + 10 = 12 terms until now ####
    N_lambda_by_T_couple_polynomial = 2
    for i in range(1,N_lambda_by_T_couple_polynomial+1):
        current_emissivity_term = lambda lambda_, i=i: ((lambda_*(10**5))/(T_couple/1000))**i
        Emissivity_terms2_symbolic.append(current_emissivity_term)

    # 5. T / lambda polynomial terms #### 2 + 12 = 14 terms until now ####
    N_T_by_lambda_couple_polynomial = 2
    for i in range(1,N_T_by_lambda_couple_polynomial+1):
        current_emissivity_term = lambda lambda_, i=i: ((T_couple/1000)/(lambda_*(10**5)))**i
        Emissivity_terms2_symbolic.append(current_emissivity_term)

    # 6. 1/lambda polynomial terms #### 2 + 14 = 16 terms until now ####
    N_1_by_lambda_polynomial = 2
    for i in range(1,N_1_by_lambda_polynomial+1):
        current_emissivity_term = lambda lambda_, i=i: ((1/(lambda_*(10**5)))**i) * (T_couple/T_couple)
        Emissivity_terms2_symbolic.append(current_emissivity_term)

    # 7. 1/T polynomial terms #### 2 + 16 = 18 terms until now ####
    N_1_by_T_couple_polynomial = 2
    for i in range(1,N_1_by_T_couple_polynomial+1):
        current_emissivity_term = lambda lambda_, i=i: (1/(T_couple/1000))**i
        Emissivity_terms2_symbolic.append(current_emissivity_term)

    # 8. 1/(lambda * T) polynomial terms #### 2 + 18 = 20 terms until now ####
    N_1_by_lambda_T_couple_polynomial = 2
    for i in range(1,N_1_by_lambda_T_couple_polynomial+1):
        current_emissivity_term = lambda lambda_, i=i: (1/((lambda_*(10**5))*(T_couple/1000)))**i
        Emissivity_terms2_symbolic.append(current_emissivity_term)

    ############################################
    #### Now let's do some fractional polynomials like 1/2 and 3/2 powers for all the above
    ############################################
    # 9. lambda fractional polynomial terms #### 2 + 20 = 22 terms until now ####
    N_lambda_fp = 2
    # this means two terms 1/2 and 3/2 of the form i - 1/2, or (2*i -1)/2 for i {1,2} in this case 
    for i in range(1,N_lambda_fp+1):
        current_emissivity_term = lambda lambda_, i=i: ((lambda_*(10**5))**(i-0.5)) * (T_couple/T_couple)
        Emissivity_terms2_symbolic.append(current_emissivity_term)

    # 10. T fractional polynomial terms #### 2 + 22 = 24 terms until now ####
    N_T_couple_fp = 2
    for i in range(1,N_T_couple_fp+1):
        current_emissivity_term = lambda lambda_, i=i: (T_couple/1000)**(i-0.5)
        Emissivity_terms2_symbolic.append(current_emissivity_term)

    # 11. lambda * T fractional polynomial terms #### 2 + 24 = 26 terms until now ####
    N_lambda_T_couple_fp = 2
    for i in range(1,N_lambda_T_couple_fp+1):
        current_emissivity_term = lambda lambda_, i=i: ((lambda_*(10**5))*(T_couple/1000))**(i-0.5)
        Emissivity_terms2_symbolic.append(current_emissivity_term)

    # 12. lambda / T fractional polynomial terms #### 2 + 26 = 28 terms until now ####
    N_lambda_by_T_couple_fp = 2
    for i in range(1,N_lambda_by_T_couple_fp+1):
        current_emissivity_term = lambda lambda_, i=i: ((lambda_*(10**5))/(T_couple/1000))**(i-0.5)
        Emissivity_terms2_symbolic.append(current_emissivity_term)

    # 13. T / lambda fractional polynomial terms #### 2 + 28 = 30 terms until now ####
    N_T_by_lambda_couple_fp = 2
    for i in range(1,N_T_by_lambda_couple_fp+1):
        current_emissivity_term = lambda lambda_, i=i: ((T_couple/1000)/(lambda_*(10**5)))**(i-0.5)
        Emissivity_terms2_symbolic.append(current_emissivity_term)

    # 14. 1/lambda fractional polynomial terms #### 2 + 30 = 32 terms until now ####
    N_1_by_lambda_fp = 2
    for i in range(1,N_1_by_lambda_fp+1):
        current_emissivity_term = lambda lambda_, i=i: ((1/(lambda_*(10**5)))**(i-0.5)) * (T_couple/T_couple)
        Emissivity_terms2_symbolic.append(current_emissivity_term)

    # 15. 1/T fractional polynomial terms #### 2 + 32 = 34 terms until now ####
    N_1_by_T_couple_fp = 2
    for i in range(1,N_1_by_T_couple_fp+1):
        current_emissivity_term = lambda lambda_, i=i: (1/(T_couple/1000))**(i-0.5)
        Emissivity_terms2_symbolic.append(current_emissivity_term)

    # 16. 1/(lambda * T) fractional polynomial terms #### 2 + 34 = 36 terms until now ####
    N_1_by_lambda_T_couple_fp = 2
    for i in range(1,N_1_by_lambda_T_couple_fp+1):
        current_emissivity_term = lambda lambda_, i=i: (1/((lambda_*(10**5))*(T_couple/1000)))**(i-0.5)
        Emissivity_terms2_symbolic.append(current_emissivity_term)

    ############################################
    #### Now let's do some exponentials without inner coefficients
    ############################################
    # 17. lambda exponential terms #### 2 + 36 = 38 terms until now ####
    N_lambda_exp = 2
    for i in range(1,N_lambda_exp+1):
        current_emissivity_term = lambda lambda_, i=i: (np.exp((lambda_*(10**5))**(i))) * (T_couple/T_couple)
        Emissivity_terms2_symbolic.append(current_emissivity_term)

    # 18. T exponential terms #### 2 + 38 = 40 terms until now ####
    N_T_couple_exp = 2
    for i in range(1,N_T_couple_exp+1):
        current_emissivity_term = lambda lambda_, i=i: smg.exp((T_couple/1500)**(i))
        Emissivity_terms2_symbolic.append(current_emissivity_term)

    # 19. lambda * T exponential terms #### 2 + 40 = 42 terms until now ####
    N_lambda_T_couple_exp = 2
    for i in range(1,N_lambda_T_couple_exp+1):
        current_emissivity_term = lambda lambda_, i=i: smg.exp(((lambda_*(10**5))*(T_couple/1500))**(i))
        Emissivity_terms2_symbolic.append(current_emissivity_term)

    # 20. lambda / T exponential terms #### 2 + 42 = 44 terms until now ####
    # N_lambda_by_T_couple_exp = 2
    # Term 44 = 43 + 1, was identified as bad.
    # Could do the scaling of T and lambda separately for this one term.
    N_lambda_by_T_couple_exp = 1
    for i in range(1,N_lambda_by_T_couple_exp+1):
        current_emissivity_term = lambda lambda_, i=i: smg.exp(((lambda_*(10**5))/(T_couple/1500))**(i))
        Emissivity_terms2_symbolic.append(current_emissivity_term)

    # 21. T / lambda exponential terms #### 2 + 44 = 46 terms until now ####
    N_T_by_lambda_couple_exp = 2
    for i in range(1,N_T_by_lambda_couple_exp+1):
        current_emissivity_term = lambda lambda_, i=i: smg.exp(((T_couple/1500)/(lambda_*(10**5)))**(i))
        Emissivity_terms2_symbolic.append(current_emissivity_term)

    # 22. 1/lambda exponential terms #### 2 + 46 = 48 terms until now ####
    N_1_by_lambda_exp = 2
    for i in range(1,N_1_by_lambda_exp+1):
        current_emissivity_term = lambda lambda_, i=i: (np.exp((1/(lambda_*(10**5)))**(i))) * (T_couple/T_couple)
        Emissivity_terms2_symbolic.append(current_emissivity_term)

    # 23. 1/T exponential terms #### 2 + 48 = 50 terms until now ####
    # N_1_by_T_couple_exp = 2
    # Term 50 = 49 + 1 was identifed as bad.
    # Could do the scaling of T and lambda separately for this one term.
    N_1_by_T_couple_exp = 1
    for i in range(1,N_1_by_T_couple_exp+1):
        current_emissivity_term = lambda lambda_, i=i: smg.exp((1/(T_couple/1500))**(i))
        Emissivity_terms2_symbolic.append(current_emissivity_term)

    # 24. 1/(lambda * T) exponential terms #### 2 + 50 = 52 terms until now ####
    # N_1_by_lambda_T_couple_exp = 2
    # Term 52 = 51 + 1 was identifed as bad.
    # Could do the scaling of T and lambda separately for this one term.
    N_1_by_lambda_T_couple_exp = 1
    for i in range(1,N_1_by_lambda_T_couple_exp+1):
        current_emissivity_term = lambda lambda_, i=i: smg.exp((1/((lambda_*(10**5))*(T_couple/1500)))**(i))
        Emissivity_terms2_symbolic.append(current_emissivity_term)

    ############################################
    #### Now let's do some natural logarithm without inner coefficients
    ############################################
    # 25. lambda ln terms #### 2 + 52 = 54 terms until now ####
    N_lambda_ln = 2
    for i in range(1,N_lambda_ln+1):
        current_emissivity_term = lambda lambda_, i=i: (smg.log((lambda_*(5*(10**5)))**(i))) * (T_couple/T_couple)
        Emissivity_terms2_symbolic.append(current_emissivity_term)

    # Terms 55 = 54 +1 and 56 = 55 + 1 were identified as bad.
    '''
    # 26. T ln terms #### 2 + 54 = 56 terms until now ####
    N_T_couple_ln = 2
    # Term 53 = 52 + 1 was identified as bad.
    # Could do the scaling of T and lambda separately for this one term.
    for i in range(1,N_T_couple_ln+1):
    # for i in range(2,N_T_couple_ln+1):
        current_emissivity_term = lambda lambda_, i=i: smg.log((T_couple/500)**(i)))
        Emissivity_terms2_symbolic.append(current_emissivity_term)
    '''

    # Terms 57 = 56 +1 and 58 = 57 + 1 were identified as bad.
    '''
    # 27. lambda * T ln terms #### 2 + 56 = 58 terms until now ####
    N_lambda_T_couple_ln = 2
    for i in range(1,N_lambda_T_couple_ln+1):
        current_emissivity_term = lambda lambda_, i=i: smg.log(((lambda_*((10**5)))*(T_couple/500))**(i)))
        Emissivity_terms2_symbolic.append(current_emissivity_term)
    '''

    # 28. lambda / T ln terms #### 2 + 58 = 60 terms until now ####
    N_lambda_by_T_couple_ln = 2
    for i in range(1,N_lambda_by_T_couple_ln+1):
        current_emissivity_term = lambda lambda_, i=i: smg.log(((lambda_*(5*(10**5)))/(T_couple/1500))**(i))
        Emissivity_terms2_symbolic.append(current_emissivity_term)

    # Terms 61 = 60 +1 and 62 = 61 + 1 were identified as bad.
    '''
    # 29. T / lambda ln terms #### 2 + 60 = 62 terms until now ####
    N_T_by_lambda_couple_ln = 2
    for i in range(1,N_T_by_lambda_couple_ln+1):
        current_emissivity_term = lambda lambda_, i=i: smg.log(((T_couple/500)/(lambda_*((10**5))))**(i))
        Emissivity_terms2_symbolic.append(current_emissivity_term)
    '''

    # 30. 1/lambda ln terms #### 2 + 62 = 64 terms until now ####
    N_1_by_lambda_ln = 2
    for i in range(1,N_1_by_lambda_ln+1):
        current_emissivity_term = lambda lambda_, i=i: np.log((1/(lambda_*(2*(10**4))))**(i)) * (T_couple/T_couple)
        Emissivity_terms2_symbolic.append(current_emissivity_term)

    # 31. 1/T ln terms #### 2 + 64 = 66 terms until now ####
    N_1_by_T_couple_ln = 2
    for i in range(1,N_1_by_T_couple_ln+1):
        current_emissivity_term = lambda lambda_, i=i: smg.log((1/(T_couple/5000))**(i))
        Emissivity_terms2_symbolic.append(current_emissivity_term)

    # 32. 1/(lambda * T) ln terms #### 2 + 66 = 68 terms until now ####
    N_1_by_lambda_T_couple_ln = 2
    for i in range(1,N_1_by_lambda_T_couple_ln+1):
        current_emissivity_term = lambda lambda_, i=i: smg.log((1/((lambda_*(10**5))*(T_couple/5000)))**(i))
        Emissivity_terms2_symbolic.append(current_emissivity_term)

    ############################################
    #### Now let's do some sin and cos without inner coefficients
    ############################################
    # 33. lambda * sin and cos of lambda terms #### 2 + 68 = 70 terms until now ####
    current_emissivity_term = lambda lambda_: (lambda_*(2*10**5)) * np.sin((lambda_*(10**5))) * (T_couple/T_couple)
    Emissivity_terms2_symbolic.append(current_emissivity_term)

    # Term 70 deactivate to suppress absolute value
    '''
    current_emissivity_term = lambda lambda_: (lambda_*(2*10**5)) * np.abs(np.cos((lambda_*(10**5)))) * (T_couple/T_couple)
    Emissivity_terms2_symbolic.append(current_emissivity_term)
    '''
    
    # 34. T * sin and cos of T terms #### 2 + 70 = 72 terms until now ####
    current_emissivity_term = lambda lambda_: (T_couple/500) * smg.sin((T_couple/500))
    Emissivity_terms2_symbolic.append(current_emissivity_term)

    # Term 72 = 71 +1 was identified as bad.
    '''
    current_emissivity_term = lambda lambda_: (T_couple/500) * abs(smg.cos((T_couple/500))))
    Emissivity_terms2_symbolic.append(current_emissivity_term)
    '''

    # 35. (lambda * T) * sin and cos of (lambda * T) terms #### 2 + 72 = 74 terms until now ####
    current_emissivity_term = lambda lambda_: ((lambda_*(10**5)) * (T_couple/500)) * smg.sin((lambda_*(10**5)) * (T_couple/500))
    Emissivity_terms2_symbolic.append(current_emissivity_term)

    # Term 74 deactivate to suppress absolute value
    '''
    current_emissivity_term = lambda lambda_: ((lambda_*(10**5)) * (T_couple/500)) * abs(smg.cos((lambda_*(10**5)) * (T_couple/500)))
    Emissivity_terms2_symbolic.append(current_emissivity_term)
    '''

    # 36. 1/lambda * sin and cos of 1/lambda terms #### 2 + 74 = 76 terms until now ####
    current_emissivity_term = lambda lambda_: (1.0/(lambda_*(5*(10**4)))) * np.sin(1.0/(lambda_*(5*(10**4)))) * (T_couple/T_couple)
    Emissivity_terms2_symbolic.append(current_emissivity_term)

    # Term 76 deactivate to suppress absolute value
    '''
    current_emissivity_term = lambda lambda_: (1.0/(lambda_*(5*(10**4)))) * np.abs(np.cos(1.0/(lambda_*(5*(10**4))))) * (T_couple/T_couple)
    Emissivity_terms2_symbolic.append(current_emissivity_term)
    '''

    # Terms 77 = 76 +1 and 78 = 77 + 1 were identified as bad.
    '''
    # 37. 1/T * sin and cos of 1/T terms #### 2 + 76 = 78 terms until now ####
    current_emissivity_term = lambda lambda_: (1.0/(T_couple/1500)) * smg.sin(1.0/(T_couple/1500))
    Emissivity_terms2_symbolic.append(current_emissivity_term)

    current_emissivity_term = lambda lambda_: (1.0/(T_couple/1500)) * smg.cos(1.0/(T_couple/1500)))
    Emissivity_terms2_symbolic.append(current_emissivity_term)
    '''

    # Terms 79 = 78 +1 and 80 = 79 + 1 were identified as bad.
    '''
    # 38. 1/(lambda * T) * sin and cos of 1/(lambda * T) terms #### 2 + 78 = 80 terms until now ####
    current_emissivity_term = lambda lambda_: (1.0/((lambda_*(10**5)) * (T_couple/1500))) * smg.sin(1.0/((lambda_*(10**5)) * (T_couple/1500)))
    Emissivity_terms2_symbolic.append(current_emissivity_term)

    current_emissivity_term = lambda lambda_: (1.0/((lambda_*(10**5)) * (T_couple/1500))) * smg.cos(1.0/((lambda_*(10**5)) * (T_couple/1500)))
    Emissivity_terms2_symbolic.append(current_emissivity_term)
    '''

    # Terms 81 = 80 +1 and 82 = 81 + 1 were identified as bad.
    '''
    # 39. lambda / T * sin and cos of lambda / T terms #### 2 + 80 = 82 terms until now ####
    current_emissivity_term = lambda lambda_: ((lambda_*(10**5)) / (T_couple/1500)) * smg.sin((lambda_*(10**5)) / (T_couple/1500))
    Emissivity_terms2_symbolic.append(current_emissivity_term)

    current_emissivity_term = lambda lambda_: ((lambda_*(10**5)) / (T_couple/1500)) * smg.cos((lambda_*(10**5)) / (T_couple/1500))
    Emissivity_terms2_symbolic.append(current_emissivity_term)
    '''

    # 40. T/ lambda * sin and cos of T/ lambda terms #### 2 + 82 = 84 terms until now ####
    current_emissivity_term = lambda lambda_: ((T_couple/1500) / (lambda_*(10**5))) * smg.sin((T_couple/1500) / (lambda_*(10**5)))
    Emissivity_terms2_symbolic.append(current_emissivity_term)

    current_emissivity_term = lambda lambda_: ((T_couple/1500) / (lambda_*(10**5))) * smg.cos((T_couple/1500) / (lambda_*(10**5)))
    Emissivity_terms2_symbolic.append(current_emissivity_term)

    ############################
    # 41. Polymonial * exponential: 5 * 2 = 10 terms following 
    #### 84 + 10 = 94 terms until now ####
    # lambda *(exp T)
    current_emissivity_term = lambda lambda_: (lambda_*(10**5)) * smg.exp((T_couple/1500))
    Emissivity_terms2_symbolic.append(current_emissivity_term)
    # lamda *exp (lambda *T)
    current_emissivity_term = lambda lambda_: (lambda_*(10**5)) * smg.exp((lambda_*(10**5)) * (T_couple/1500))
    Emissivity_terms2_symbolic.append(current_emissivity_term)
    # lamda *exp (lambda /T)
    current_emissivity_term = lambda lambda_: (lambda_*(10**5)) * smg.exp((lambda_*(10**5)) / (T_couple/1500))
    Emissivity_terms2_symbolic.append(current_emissivity_term)
    # lamda *exp (T/lambda)
    current_emissivity_term = lambda lambda_: (lambda_*(10**5)) * smg.exp((T_couple/1500) / (lambda_*(10**5)))
    Emissivity_terms2_symbolic.append(current_emissivity_term)
    # lamda *exp (1/(lambda *T))
    current_emissivity_term = lambda lambda_: (lambda_*(10**5)) * smg.exp(1.0/((lambda_*(10**5)) * (T_couple/1500)))
    Emissivity_terms2_symbolic.append(current_emissivity_term)

    # T *(exp lambda)
    current_emissivity_term = lambda lambda_: (T_couple/1500) * np.exp((lambda_*(10**5)))
    Emissivity_terms2_symbolic.append(current_emissivity_term)
    # T *exp (lambda *T)
    current_emissivity_term = lambda lambda_: (T_couple/1500)  * smg.exp((lambda_*(10**5)) * (T_couple/1500))
    Emissivity_terms2_symbolic.append(current_emissivity_term)
    # T *exp (lambda /T)
    current_emissivity_term = lambda lambda_: (T_couple/1500)  * smg.exp((lambda_*(10**5)) / (T_couple/1500))
    Emissivity_terms2_symbolic.append(current_emissivity_term)
    # T *exp (T/lambda)
    current_emissivity_term = lambda lambda_: (T_couple/1500)  * smg.exp((T_couple/1500) / (lambda_*(10**5)))
    Emissivity_terms2_symbolic.append(current_emissivity_term)
    # T *exp (1/(lambda *T))
    current_emissivity_term = lambda lambda_: (T_couple/1500)  * smg.exp(1.0/((lambda_*(10**5)) * (T_couple/1500)))
    Emissivity_terms2_symbolic.append(current_emissivity_term)

    ############################
    # 42. Polymonial * ln: 6 terms following 
    #### 94 + 6 = 100 terms until now ####
    # Terms 95 = 94 +1 and 96 = 95 + 1 were identified as bad.
    '''
    # lamda *( ln T)
    current_emissivity_term = lambda lambda_: (lambda_*(10**5)) * smg.log((T_couple/500))
    Emissivity_terms2_symbolic.append(current_emissivity_term)
    
    # lamda *ln (lambda *T)
    current_emissivity_term = lambda lambda_: (lambda_*(10**5)) * smg.log((lambda_*(10**5)) * (T_couple/500))
    Emissivity_terms2_symbolic.append(current_emissivity_term)
    '''
    
    # lamda *ln (1/(lambda *T))
    current_emissivity_term = lambda lambda_: (lambda_*(10**5)) * smg.log(1.0/((lambda_*(10**5)) * (T_couple/5000)))
    Emissivity_terms2_symbolic.append(current_emissivity_term)
    
    # T *(ln lambda)
    current_emissivity_term = lambda lambda_: (T_couple/500) * np.log((lambda_*(2*(10**5))))
    Emissivity_terms2_symbolic.append(current_emissivity_term)

    # Term 99 = 98 +1 was identified as bad.
    '''
    # T *ln (lambda *T)
    current_emissivity_term = lambda lambda_: (T_couple/500) * smg.log((lambda_*(10**5)) * (T_couple/500))
    Emissivity_terms2_symbolic.append(current_emissivity_term)
    '''
    # T *ln (1/(lambda *T))
    current_emissivity_term = lambda lambda_: (T_couple/1500)  * smg.log(1.0/((lambda_*(2*(10**4))) * (T_couple/1500)))
    Emissivity_terms2_symbolic.append(current_emissivity_term)

    # Terms [43, 49, 51, 54, 55, 56, 57, 60, 61, 71, 76, 77, 78, 79, 80, 81, 94, 95, 98] were identified as bad,
    # Starting at zero so add 1 to all terms
    
    
    #1st dorpping of terms
    negative_coefficient_indices = [47,48,69,67]

    '''
    np.array(Emissivity_term_coeffs3)[np.array([47,48,69,67])]

    array(['a_1exp_1/T', 'a_1exp_1/lT', 'a_l_exp_1/lT', 'a_l_exp_l/T'],
          dtype='<U13')

    # Term 67 contributes to negative coefficient in the Right Thermocouple 20 cm min-1 experiment and
    terms 47,48,69 contribute to negative coefficient in the Left Thermocouple 20 cm min-1 experiment
    '''

    mask = np.ones(np.shape(Emissivity_terms2_symbolic), bool)
    mask[negative_coefficient_indices] = False
    Emissivity_terms3_symbolic = list(np.array(Emissivity_terms2_symbolic)[mask])
    
    # 2nd dropping of terms
    negative_coefficient_indices = [11,42]

    '''
    array(['a_2_l/T', 'a_1exp_l/T'], dtype='<U13')

    # Terms 11 and 42 after having already removed 4 terms contribute to negative coefficients in Left Thermocouple 20 cm min-1
    experiment
    and term 42 also contributes to negative coefficient in Right Thermocouple 20 cm min-1 experiment
    '''
    
    
    mask = np.ones(np.shape(Emissivity_terms3_symbolic), bool)
    mask[negative_coefficient_indices] = False
    Emissivity_terms3_symbolic = list(np.array(Emissivity_terms3_symbolic)[mask])
    
    
    #3rd dropping of terms
    negative_coefficient_indices = [68,66]
    '''
    array(['a_T_exp_1/lT', 'a_T_exp_l/T'], dtype='<U13')

    # Terms 68 and 66 contribute to negative coefficients in Left and Right thermocouple 20 cm min-1 experiment
    '''
    
    mask = np.ones(np.shape(Emissivity_terms3_symbolic), bool)
    mask[negative_coefficient_indices] = False
    Emissivity_terms3_symbolic = list(np.array(Emissivity_terms3_symbolic)[mask])
    
    # 4th dropping of terms
    negative_coefficient_indices = [16, 18, 36, 38, 40, 42, 46, 48, 62, 63, 65, 66, 35, 43, 44, 50, 52, 54]
    '''
    array(['a_2_1/T', 'a_2_1/lT', 'a_2exp_l', 'a_2exp_T', 'a_2exp_lT',
           'a_2exp_T/l', 'a_2ln_l', 'a_2ln_l/T', 'a_l_exp_lT', 'a_l_exp_T/l',
           'a_T_exp_lT', 'a_T_exp_T/l', 'a_1exp_l', 'a_1exp_1/l',
           'a_2exp_1/l', 'a_2ln_1/l', 'a_2ln_1/T', 'a_2ln_1/lT'], dtype='<U13')
    # Terms that had negative coefficients in Left and Right 10 cm min-1 experiemnt
    '''
    mask = np.ones(np.shape(Emissivity_terms3_symbolic), bool)
    mask[negative_coefficient_indices] = False
    Emissivity_terms3_symbolic = list(np.array(Emissivity_terms3_symbolic)[mask])
    
    # 5th dropping of terms
    negative_coefficient_indices =  [9,12,37,39,40,49]
    '''
    array(['a_2_lT', 'a_2_T/l', 'a_1ln_l/T', 'a_1ln_1/T', 'a_1ln_1/lT',
           'a_l_ln_1/lT'], dtype='<U13')
    # Terms 9 and 12 had negative coefficients in Left thermocouple and 
    terms 37,39,40,49 had negative coefficients in Right thermocouple, 10 cm min-1 experiments
    '''
    mask = np.ones(np.shape(Emissivity_terms3_symbolic), bool)
    mask[negative_coefficient_indices] = False
    Emissivity_terms3_symbolic = list(np.array(Emissivity_terms3_symbolic)[mask])
    
    # 6th dropping of terms
    negative_coefficient_indices =  [7]
    '''
    array(['a_2_T'], dtype='<U13') 
    # Terms 7 had negative coefficients in Left thermocouple 10 cm min-1 experiment
    '''
    mask = np.ones(np.shape(Emissivity_terms3_symbolic), bool)
    mask[negative_coefficient_indices] = False
    Emissivity_terms3_symbolic = list(np.array(Emissivity_terms3_symbolic)[mask])
    
    # 7th dropping of terms
    negative_coefficient_indices =  [21,23]
    '''
    array(['a_3/2_l/T', 'a_3/2_T/l'], dtype='<U13')
    # Terms 21 and 23 have negative coefficients in Left thermocouple 10 cm min-1 experiment
    '''
    mask = np.ones(np.shape(Emissivity_terms3_symbolic), bool)
    mask[negative_coefficient_indices] = False
    Emissivity_terms3_symbolic = list(np.array(Emissivity_terms3_symbolic)[mask])
    
    #Emissivity_terms3_symbolic = Emissivity_terms2_symbolic
    return Emissivity_terms3_symbolic
```

```python
def integrator_with_emissivity_term_s_Array_w_responsivity_symbolic(B_lambda_T, responsivity_tuples, Emissivity_terms):
    '''
    Arguments:
    1. B_lambda_T : is a lambda function of wavelength and returns an expression in the symbolic variable T_couple
    
    3. One_emissivity_term: a lambda function that takes in a wavelength and returns an expression in 
    the symbolic variable T_couple
            This is an element in the list which is the return value of the function 
            Emissivity_terms2_symbolic(T_couple)

    Returns: 
    An sympy matrix of elements that are the numerical integration of one of the emissivity term, as an expression in T_couple
    '''
    # Get the number of terms of the emissivity function
    number_of_emissivity_terms = np.size(Emissivity_terms)
    
    # This is an array to hold the integrated terms that are expressions of T_couple
    Array_integrated_non_blackbody_intensity = []

    
    # For each term of the emissivity model
    for i in range(number_of_emissivity_terms):
        
        # Append the expression for blackbody intensity and the emissivity to the list of functions to be multiplied
        # Both are lambda functions of wavelength at the same temperature
        functions_to_be_multiplied = [B_lambda_T, Emissivity_terms[i]]
        
        # Call the trapezoidal function and get the result of integration and put it into the array
        Array_integrated_non_blackbody_intensity.append(trapezoidal_multiplier_w_responsivity(functions_to_be_multiplied, responsivity_tuples))
        
        print(f'{i+1} of {number_of_emissivity_terms} done.')
        
    return smp.Matrix(np.array(Array_integrated_non_blackbody_intensity))
```

```python
def Array_of_expressions_of_T_function(T_couple, responsivity_tuples):
    
    # Get the blackbody B and emissivity terms as expressions of T_couple
    B_lambda_T = B_lambda_T_at_T_symbolic(T_couple)
    
    # To change the emissivity model will need to define a new function and replace this line
    Emissivity_terms = Emissivity_terms2_symbolic(T_couple)
    
    Array_of_expressions_of_T = integrator_with_emissivity_term_s_Array_w_responsivity_symbolic(B_lambda_T, responsivity_tuples, Emissivity_terms)
    
    return Array_of_expressions_of_T
```

```python
def Row_emissivity_normalizer_2(B_0_list):
    B_0_0 = B_0_list[0]
    B_0_divided = B_0_list / B_0_0
    B_0_normalized = B_0_divided / np.sum(B_0_divided)
    B_0_norm_multiplied = B_0_normalized * B_0_0
    
    return B_0_norm_multiplied
```

```python
def secant(B_int_blackbody, Coefficients_of_Emissivity_terms, the_expression_of_T, T_1, N_iterations, T_couple):
    # N_iterations is the number of iterations
    # B_int_of_T is an expression of T_couple -- the expression we get by having done integral
    # T_1 will be T cam
    
    T_0 = T_1 - 5
    
    first_time = datetime.datetime.now()
    #B_0 = B_int_of_T.subs({T_couple:T_0})
    
    '''
    ###################################
    ###################################
    #B_0_list = the_expression_of_T(T_0)
    ##### For row emissivity normalization
    #######################################
    #####################################
    #####################################
    B_0_list = np.array(the_expression_of_T(T_0), dtype='float')
    B_0_list_norm = Row_emissivity_normalizer_2(B_0_list)
    ######################################
    #####################################
    #####################################
    '''
    
    B_0_list = the_expression_of_T(T_0)
    B_0 = float(np.dot(B_0_list, Coefficients_of_Emissivity_terms))
    #B_0 = B_int_of_T.xreplace({T_couple:T_0})
    #B_0 = float(B_0)
    second_time = datetime.datetime.now()
    print(f'Substituting T_0 to get B_0 took: {second_time - first_time}')
    
    '''
    # Use if necessary
    if B_0 == float("inf"):
        print(f'B_0 is infinity')
        return
    '''
    
    for i in range(N_iterations):
        third_time = datetime.datetime.now()
        #B_1 = B_int_of_T.subs({T_couple:T_1})
        #B_1 = B_int_of_T.xreplace({T_couple:T_1})
        #B_1 = float(B_1)
        
        '''
        ##################################
        #####################################
        #B_1_list = the_expression_of_T(T_1)
        #####################################
        ######################################
        #######################################
        ##### For row emissivity normalization
        B_1_list = np.array(the_expression_of_T(T_1), dtype='float')
        B_1_list_norm = Row_emissivity_normalizer_2(B_1_list)
        ######################################
        ####################################
        ####################################
        '''
        
        B_1_list = the_expression_of_T(T_1)
        B_1 = float(np.dot(B_1_list, Coefficients_of_Emissivity_terms))
        fourth_time = datetime.datetime.now()
        print(f'Substituting T_1 to get B_1 took: {fourth_time - third_time}')
        '''
        # Use if necessary
        if B_1 == float("inf"):
        print(f'B_1 is infinity')
        return
        '''
        
        # So that we don't divide by zero
        if B_1 == B_0:
            print(f'It took: {i} iterations to converge to a value of {T_1}')
            break
            
        T_2 = T_1 - (B_1 * (T_1 - T_0)/(B_1 - B_0)) + (B_int_blackbody * (T_1 - T_0)/(B_1 - B_0))
        
        '''
        # Use if the value stuck around the same point for fairly long.
        I think might already be checked because we are checking for B_1 == B_0 earlier
        if np.isclose(T_2, T_1, atol=1e-3, rtol=0):
            print(f'It took: {i} iterations to converge to a value of {T_2}')
            break
        '''

        T_0 = T_1
        # Also copy the value of B so we don't have to evaluate it again
        B_0 = B_1
        
        T_1 = T_2
        
        print(f'T_2 now: {T_2}')
    
    return T_2
```

```python
def get_Corrected_Temperature_s(B_int_blackbody_s_Test, TCam_s_Test, Coefficients_of_Emissivity_terms, an_Array_of_expressions_of_T, N_iterations, T_couple):
    
    N_temperatures = np.shape(TCam_s_Test)[0]
    
    T_s_corrected = []
    
    # First get the expression of T
    start_time = datetime.datetime.now()
    #the_expression_of_T = (an_Array_of_expressions_of_T[:5]).dot(np.array(Coefficients_of_Emissivity_terms)[:5])
    end_time = datetime.datetime.now()
    the_expression_of_T = smg.lambdify(T_couple, an_Array_of_expressions_of_T)
    print(f'The lambdifying takes: {end_time - start_time}')
    
    for i in range(N_temperatures):
        start_time = datetime.datetime.now()
        B_int_blackbody = B_int_blackbody_s_Test[i]
        T_1 = TCam_s_Test[i]
        
        T_corrected = secant(B_int_blackbody, Coefficients_of_Emissivity_terms, the_expression_of_T, T_1, N_iterations, T_couple)
        
        T_s_corrected.append(T_corrected)
        
        if i%10 == 0:
            print(f'Frame {i} of {N_temperatures} finished at {datetime.datetime.now()}')
        
        end_time = datetime.datetime.now()
        print(f'This frame took: {end_time-start_time}')
    return np.array(T_s_corrected)
```
